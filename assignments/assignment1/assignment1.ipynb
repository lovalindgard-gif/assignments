```python
#Task 1:

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

#a)

valid_transactions = [t for t in transactions if t["price"] > 0 and t["quantity"] > 0]

print(valid_transactions)

#b)

for t in valid_transactions:
    t["total"]= t["price"] * t["quantity"]
    print(t)

#c)

revenue_by_region = {}

for t in valid_transactions:
    region = t["region"]
    total = t["total"]

    if region not in revenue_by_region:
        revenue_by_region[region] = 0
    revenue_by_region[region] += total

print("Revenue by region", revenue_by_region)

#d)

highest_region = max(revenue_by_region, key=revenue_by_region.get)
highest_revenue = revenue_by_region[highest_region]

print(f"The region with the highest total revenue is {highest_region} with ${highest_revenue}")
```

```python
#Task 2:

sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales (sales):
    total = sum (sales)
    average = sum (sales) / len (sales)
    best_day = sales.index(max(sales)) + 1
    worst_day = sales.index(min(sales)) + 1
    above_average_count = len ([day for day in sales if day > average])
    return {
        "total": total,
        "average" : average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count

     }

result = summarise_sales(sales)
print(result)
```

```python
#Task 4:
#a) 
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }

    def total_value(self) :
        return sum(
        data["shares"] * data["price_per_share"]
        for data in self.holdings.values()
    )
    
    def most_valuable(self):
        return max(
        self.holdings,
        key = lambda ticker: self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"]
    )
    
    def __str__(self):
        return(
        f"Portfolio Owner : {self.owner}\n"
        f"Number of Assets: {len(self.holdings)}\n"
        f"Total Value     : ${self.total_value():,.2f}"
)


# === Demonstration ===
p = Portfolio("Leia")
p.add_asset("NVDA",  4, 912.50)
p.add_asset("MSFT",  3, 456.70)
p.add_asset("GOOGL", 9, 286.77)

print("=== Initial Portfolio ===")
print(p)

print("\n=== Most Valuable Asset ===")
print(f"Ticker: {p.most_valuable()}")

print("\n=== Add More GOOGL Shares (price update) ===")
p.add_asset("GOOGL", 3, 296.30)
googl = p.holdings["GOOGL"]
print(f"GOOGL: {googl['shares']} shares @ ${googl['price_per_share']}")
print(f"New Total Value: ${p.total_value():,.2f}")

      
```

```python
#Task 5:

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

from datetime import date

dates = [date.fromisoformat(order[0]) for order in orders] #This parses each date into a date object

earliest = min(dates)
latest = max(dates)  #This gives us the earliest and latest dates

difference = latest - earliest #Number of days between earliest and latest order


print(f"Earliest order date: {earliest}")
print(f"Latest order date: {latest}")
print(f"Days between these orders: {difference.days}")
```

```python
#Task 6:


import numpy as np 


# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)


revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])



#a) 
total_revenue = np.sum (revenue, axis = 1)

#b)
mean_revenue = np.mean (revenue, axis = 0)

#I'm adding labels to the years and quarters for better readability as well as rounding up the decimals to 2 decimals

print(f"Total revenue for each year:")

years =[2022, 2023, 2024]
for year, total in zip(years, total_revenue):
    print(f"{year}: {total}")
    

print(f"Median revenue for each quarter:")

quarters = ["Q1", "Q2", "Q3", "Q4"]
for quarter, mean in zip(quarters, mean_revenue):
    print(f"{quarter}:{mean:.2f}")


#c)

growth_adjustment = revenue[2] * [1.05, 1.03, 1.08, 1.02]

print(f"Growth adjustment for the quarters of 2024:")

for quarter, growth in zip(quarters, growth_adjustment):
    print(f"{quarter}:{growth:.2f}")


#d)

mask = revenue > 250  #kom tillbaka hit

revenue[mask]

print(f"Quarterly revenues that exceed 250 (thousand euros): {revenue[mask]}")
```

```python
#Task 8:

import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

#a)

data = {
    "date" : days,
    "visitors" : visitors,
    "conversion_rate" : conversion_rate,
    "avg_order_value" : avg_order_value
}

df = pd.DataFrame(data).set_index("date")
print(df)

#b) 

df.insert(3,"conversions", (visitors * conversion_rate).round(), True)

conversions = visitors * conversion_rate

df.insert(4, "daily_revenue", (conversions * avg_order_value), True)

print(df)
    
```
